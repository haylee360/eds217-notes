## Exploring Banana Scores
For each of the pre-computed banana score columns (kg, calories, and protein), show the 10 highest-scoring food products.

Edit the function below so that is returns the top 10 scores for a given column:

## Common high scoring foods

```python
df.sort_values('Bananas index (kg)', ascending = False).head(10)
df.sort_values('Bananas index (1000 kcalories)', ascending = False).head(10)
df.sort_values('Bananas index (100g protein)', ascending = False).head(10)
```

```python
# Create a function to sort 
def return_top_ten(df, column):
    new_df = df.sort_values(column, ascending = False).head(10).copy()
    return new_df[column]
```

```python
# Assign to df
kg_df = return_top_ten(df, column='Bananas index (kg)')
kcal_df = return_top_ten(df, column='Bananas index (1000 kcalories)')
protein_df = return_top_ten(df, column='Bananas index (100g protein)')
print(protein_df)
kg_df.dtypes
```

```python
# change to set
kg_set = set(kg_df.index)
kcal_set = set(kcal_df.index)
protein_set = set(protein_df.index)
print(kg_set)
print(kcal_set)
print(protein_set)
```

```python
# find intersection
in_all_three = set.intersection(kg_set, kcal_set, protein_set)
print(in_all_three)
```
